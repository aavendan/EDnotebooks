### Código en Java

1. En una celda vacía, escribe el código en **Java**.
2. Agrega al inicio la instrucción `%%writefile`, seguido del nombre del archivo con extensión `.java`
3. Ejecute la celda con el código.


```python
# Ejecute esta celda

%%writefile HelloWorld.java

public class HelloWorld {
    public static void main(String[] args) {
        System.out.println("Hola desde Java en Google Colab!");
    }
}
```

### Archivo con extensión java

### Compile el código

1. En una celda vacía, escriba la instrucción `!javac`, acompañada por el nombre del archivo con extensión `java`.
2. Ejecute la celda.

```python
# Ejecute esta celda

!javac HelloWorld.java
```
